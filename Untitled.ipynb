```python
!pip install keras
!pip install tensorflow
import pandas as pd 
import numpy as np
from keras.models import Sequential
from keras.layers import Dense,Dropout
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv'

columns =['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction',
          'Age','Outcome']
df = pd.read_csv(url,header=None,names = columns)
```

```python
df.head()
```

```python
x = df.drop('Outcome',axis = 1)
```

```python
y = df['Outcome']
```

```python
for col in x:
    x[col].fillna(x[col].median(),inplace = True)
```

```python
df.Outcome.value_counts()
```

```python
scaler = StandardScaler()
```

```python
x_scaled=scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(
    x_scaled,
    y,
    test_size = 0.2,
    random_state = 42
)
```

```python
model = Sequential() #Brain
model.add(Dense(64,input_dim=8,activation = 'relu')) #First Hidden Layer
model.add(Dense(32,activation = 'relu'))#Second Hideen Layer
#model.add(Dense(64,activation = 'relu')) #Third Hidden Layer
#model.add(Dense(32,activation = 'relu')) #Fourth Hideen Layer
model.add(Dense(16,activation = 'relu')) #Fifth Hideen Layer
model.add(Dense(8,activation = 'relu')) #Sixth HidDEN Layer
model.add(Dense(1,activation = 'sigmoid')) #Output Layer of Binary Classification
```

```python
#COMPILE THE MODEL 
model.compile(optimizer='adam',loss='binary_crossentropy',metrics = ['accuracy'])
```

```python
model.fit(x_train,y_train,epochs = 20,batch_size = 128,validation_split = 0.2)
```

```python
loss,accuracy = model.evaluate(x_test,y_test)
print(f'Test Accuracy:{accuracy:.4f}')
```
